# CNN and image

- convolutional neural network and convolutional autoencoder for creating embeddings of image data


- images are a series of 2D pixel values


- main idea of conv NN: 

    because of images' built-in spatial locality structure, 
    
    we can look at full structure locally by scanning across images

# 3 basic ideas of CNN

1. local receptive fields: scan local region of images. 


2. shared weights within one conv operation: weights are entries in a filter. same weights bc a filter detect one feature. reduce number of params to faster training. 


3. pooling: summarize feature to reduce complexity. type: max, average.

# convolution operation

- a filter/feature detector/kernel: used to search for a particular feature (from very low level to very high level)
    
    e.g., Gaussian blur (approximation), box blur (normalized), sharpen, edge detection, identity
    
    value of a filter is weight for local receptive field


- features of filter

    - pointwise multiply
    
    - add results
    
    - translate filter


- stride: controls how many pixels of an image to slide over

- padding: zero padding is padding 0s to outside of an image

- depth = number of filters used = number of features maps

# Convolutional AutoEncoder (CAE)

- conv Autoencoder has similar architecture as vanilla Autoencoder


- input image > conv layers and pooling layers (encoding, reduce dim) > latent space representation > transpose conv and upsampling
(decoding, increase dim) > reconstructed image

## decoding

### transposing convolutions 

the transpose convolution operation has nothing to do with transpose a matrix


e.g. 3 x 3 transpose convolution, stride 2 pad 1

- input gives weight for filter


- weight gives amount of input value being distributed to each of output cells


- stride gives ratio between movement in output and input

    - filter moves 2 pixels in the output for every one pixel in the input
    
    - sum where output overlaps
    

- checkerboard artifact  : one problem with transpose convolution

    def: result of uneven overlap of transpose convolution, which make some part of output images has darker color than other part, looks like a checkerboard

### upsamling

- another method to increase dim in decoder


- used more than transpose convolution

# Representation Learning

- Machine learning: identify patterns in data

    2 types of ML: 

    - supervised learning: given a bunch of labelled data, want to label new data

        learn a function $f(X) \rightarrow Y$ where all values of $Y$ are known for some samples of $X$

    - unsupervised learning: given a bunch of unlabelled data, want to organize them

        learn an embedding $f(X): X \in \mathbb{R}^{n}  \mapsto Y \in \mathbb{R}^{m} (n \gg m)$ where $Y$ is in lower dimension, easier to interpret

- big data: many observations (high throughput, volumn), high dimensional (many measurements), heterogeneous (variety)

- representation learning/dimensionality reduction methods:

    used for different dataset contexts.

    PCA, MDS

    t-SNE, UMAP

    kernel PCA, Diffusion Maps, PHATE

## spatial construction

- Apply the same operator and move over the graph space (vector space of a graph) in some fashion


- notion of locality based on neighbors of a node

### GraphSAGE network (Hamilton, 2017)

### aggregation function

1. mean aggregation: most common aggregation function, take mean of features on neighbor nodes
$$
h_v^k \leftarrow \sigma \left ( W \cdot \text{Mean}\left ( \left\{ h_v^{k-1} \right\}\cup \left\{ h_u^{k-1}, \forall u \in N(v) \right\} \right )\right )
$$

2. max-pooling aggregation: take max value of features on neighbor nodes
$$
h_v^k =\max \left ( \left\{ \sigma \left ( \textbf{W} \textbf{h}_{u_i}^k +\textbf{b} \right ), \forall u_i \in N(v) \right\} \right )
$$

3. LSTM aggregation

problem: 

- Local aggregation does not define the whole graph

4. **graph attention**: 

    
    
    $\vec{\alpha_{1i}}$: different attention weights for different neighbors
    
    $\vec{h'_1}$: weighted average of current state of node $\vec{h_1}$

## spectral construction: spectral graph convolutional network (GCN)

- define a convolutional operation in the graph spectral/Fourier domain

- notion of locality based on global properties of the filter

### summarized architecture of GCN


### Chebyshev Polynomials

- filters for GCN isn't original filters where we come up with multipliers for different Fourier coefficient


- avoid eigendecomposition which is expensive


- Chebyshev polynomials of graph Laplacian defined recursively

$$
T_k(y):= \left\{\begin{matrix}
1 & k=0  \\
y & k=1 \\
2yT_{k-1}(y) - T_{k-2}(y) & k \geq 2  \\
\end{matrix}\right.
$$

- original spectral filters

$$
y = g_{\theta} (L)x = g_{\theta} (U\Lambda U^T)x = U g_{\theta} (\Lambda) U^T x
$$


- polynomial filters: avoid eigendecomposition

$$
g_{\theta} (\Lambda)  = \sum_{k=0}^{K-1}\theta_k \Lambda ^k
$$

- Chebyshec filters: is a type of polynomial filters

$$
g_{\theta} (\Lambda)  = \sum_{k=0}^{K-1}\theta_k T_k (\tilde \Lambda)
$$


### Chebyshev Basis

- filter can just learn coefficients for Chebyshev basis


- ad: smoothness in frequency domain means they are **localized** in vertex domain

    we do lots of local operation in GNN

- Autoencoder is used to dimensionality reduction, denoising, information retrieval, anomaly detection

# vanilla Autoencoder

## architecture

- Autoencoder is a neural network trained to copy its input x to its output

- 2 parts

    - encoder: $z = f(x)$: input layer > hidden layers > code layer
    
    - decoder: $x'=g(z)$: code layer > hidden layers > output layer


- code layer z: also called **informational bottleneck**/embedding layer, learn an embedding of data 

## loss: reconstruction error

- objective: minimize reconstruction error (a kind of **self-supervised** loss) between input and reconstructed input

$$
L[x, g(f(x))]
$$

**lossy compression** : Autoencoder only approximately copies the input

# undercomplete Autoencoder

- undercomplete Autoencoder: dimension of code layer is smaller than input layer

- similar to some dim reduction methods, force Autoencoder to capture the most salient features of training data

- linear encoder + decoder + MSE loss: embedding = PCA

- nonlinear encoder and decoder: embedding like kernel PCA

# regularized Autoencoder

- **high capacity problem**: if the dim of code layer is too large, Autoencoder just copy the input to be output without extracting useful embeddings
    
- solution: use **small** dim of code layer and add **other penalty** besides reconstruction error to loss function bc we don't need a perfect reconstruction

- regularized Autoencoder add **regularization term** to the loss function to overcome the capacity problem


- regularization **plays against reconstruction** to learn meaningful embeddings

# denoising Autoencoder

- denoising Autoencoder minimize reconstruction loss of input and reconstructed **noisy** input

$$
L[x, g(f( \color{red}{\tilde x} ))]
$$

where $\tilde x$ is a noisy/corrupted version of $x$, denoising AE learn to **undo the corruption** 

- corruption process $C(\tilde x | x)$ modeled by conditional distribution


- denoising AE learns a reconstruction distribution $p(x| \tilde x)$ from training pairs $(x, \tilde x)$

## denoising and Manifold learning

- if we add some corruption to data on the Manifold,

    the corrupted data will be away from Manifold, 

    denoising AE can learn to push back corrupted data to Manifold


- black curve: data manifold


- red cross: true signal $x$


- gray circle: equiprobable corruption 等可能发生的数据损坏


- green arrow: denoising AE learn what kind of noise is **orthogonal** to manifold of data

    correction back the corrupted data $\tilde x$ back to the **nearest point** $g(f(\tilde x))$ on the manifold (reconstructed data)

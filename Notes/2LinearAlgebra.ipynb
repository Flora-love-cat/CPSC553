# Linear algebra

# transformation

- def: a transformation (function/mapping) $T$ from $\mathbb{R}^n$ to $\mathbb{R}^m$ is a rule that assigns each vector $\mathbf{x} \in \mathbb{R}^n$ to a vector $T(\mathbf{x}) \in \mathbb{R}^m$

$$
T: \mathbb{R}^n \rightarrow \mathbb{R}^m
$$

**domain** of $T$: $\mathbb{R}^n$

**codomain** of $T$: $\mathbb{R}^m$

**image** of vector $\mathbf{x}$: vector $T(\mathbf{x})$

**range** of $T$: set of all images $T(\mathbf{x})$

# matrix transformation

- def: matrix transformation is a transformation $T$ which is under rule of $T(\mathbf{x})=A\mathbf{x}$, where vector $\mathbf{x} \in \mathbb{R}^n$, $A$ is an $m \times n$ matrix

$$
T: \mathbf{x} \mapsto A \mathbf{x}
$$

$$
T: \mathbb{R}^n \rightarrow \mathbb{R}^m
$$

**domain** of $T$: $\mathbb{R}^n$

**codomain** of $T$: $\mathbb{R}^m$

**image** of vector $\mathbf{x}$: vector $T(\mathbf{x})$

**range** of $T$: set of all images $T(\mathbf{x})$, i.e. all **linear combinations** of columns of $A$

# Linear transformation

- def: a transformation $T$ is linear if:

    (i) $T(\mathbf{u}+\mathbf{v})=T(\mathbf{u})+T(\mathbf{v})\ \ $ for all $\mathbf{u}, \mathbf{v}$ in domain of $T$
    
    (ii) $T(c\mathbf{u})=cT(\mathbf{u})\ \ $ for all scalars $c$ and all $\mathbf{u}$ in domain of $T$

- properties:

    - $T(\mathbf{0})=\mathbf{0}$
    
    - $T(c\mathbf{u}+d\mathbf{v})=cT(\mathbf{u})+dT(\mathbf{v})$
    
    - superposition principle:
    
$$
T(c_1\mathbf{v_1}+...+c_p\mathbf{v_p})=c_1 T(\mathbf{v_1})+...+c_p T(\mathbf{v_p})
$$
    
    
- **every matrix transformation is a linear transformation**
  
  
- but linear transformation can have other format
    


## Stretching

- Stretch: change length of a vector without change its direction


- stretch a vector $\mathbf{x}$ by a factor of $K \in \mathbb{R}$ equals multiply the vector by a diagonal stretching matrix $A$

$$
A = \begin{bmatrix}
K & 0 \\ 
0 & K
\end{bmatrix}
$$


- Given vector $\mathbf{x}=\begin{bmatrix}
X\\ 
Y
\end{bmatrix}$, the stretched vector is $K\mathbf{x}$

$$
A\mathbf{x} = \begin{bmatrix}
K & 0 \\ 
0 & K
\end{bmatrix}\begin{bmatrix}
X\\ 
Y
\end{bmatrix}=\begin{bmatrix}
KX\\ 
KY
\end{bmatrix}=K\mathbf{x}
$$

## Rotation

- rotation: change direction of a vector without changing its length


- rotate a vector $\mathbf{x}$ by an angle of $\phi$ equals multiply the vector by a rotation matrix $A$

$$
A = \begin{bmatrix}
\cos \phi & -\sin \phi \\ 
\sin \phi & \cos \phi
\end{bmatrix}
$$

- properties of rotation matrix $A$:

    - orthonomal
    
    - can be normalized to have determinant 1


- Given vector $\mathbf{x}=\begin{bmatrix}
X\\ 
Y
\end{bmatrix}$, the rotated vector is $A\mathbf{x}$

$$
A\mathbf{x} = \begin{bmatrix}
\cos \phi & -\sin \phi \\ 
\sin \phi & \cos \phi
\end{bmatrix}\begin{bmatrix}
X\\ 
Y
\end{bmatrix}=\begin{bmatrix}
X\cos \phi  - Y \sin \phi \\ 
X \sin \phi + Y \cos \phi
\end{bmatrix}
$$

$
\left \| A\mathbf{x} \right \|=\sqrt{X^2\cos^2 \phi +Y^2\sin^2 \phi -2XY \cos\phi\sin\phi+X^2\sin^2 \phi +Y^2\cos^2 \phi +2XY \cos\phi\sin\phi}\\[1em]
=\sqrt{X^2(\sin^2+\cos^2)+Y^2(\sin^2+\cos^2)}\\[1em]
=\sqrt{X^2+Y^2}\\[1em]
=\left \| \mathbf{x} \right \|\\[1em]
$

## Shearing

- shearing: change both direction and length of a vector


- shear a vector $\mathbf{x}$ in direction of $x$ axis by a factor of $m$ proportional to its $y$ axis equals multiply the vector by a shearing matrix $A$

$$
A = \begin{bmatrix}
1 & m \\ 
0 & 1
\end{bmatrix}
$$


- Given vector $\mathbf{x}=\begin{bmatrix}
X\\ 
Y
\end{bmatrix}$, the sheared vector is $A\mathbf{x}$

$$
A\mathbf{x} = \begin{bmatrix}
1 & m \\ 
0 & 1
\end{bmatrix}\begin{bmatrix}
X\\ 
Y
\end{bmatrix}=\begin{bmatrix}
X+{\color{Red} m}Y \\ 
Y
\end{bmatrix}
$$

- shearing vector $\mathbf{x}=\begin{bmatrix}
1\\ 
1
\end{bmatrix}$ with $m=3$, the sheared vector is $\begin{bmatrix}
4\\ 
1
\end{bmatrix}$

- shearing a vector $\mathbf{x}=\begin{bmatrix}
1\\ 
0
\end{bmatrix}$ with $m=3$, **doesn't change the vector** 


- reason: $\mathbf{x}$ is an eigenvector of matrix $A$


$$
A = \begin{bmatrix}
1 & 3 \\ 
0 & 1
\end{bmatrix}
$$

```python
import numpy as np
from matplotlib import pyplot as plt

K=3  # stretching factor
phi= -np.pi/4 # rotate angle
m = 3 # shearing factor
x = np.array([[1],[1]])
A1 = np.array([[K,0],[0,K]])  # a diagonal stretching matrix
A2 = np.array([[np.cos(phi), -np.sin(phi)],[np.sin(phi),np.cos(phi)]]) # rotation matrix
A3 = np.array([[1, m],[0, 1]]) # shearing matrix
x1 = A1 @ x # stretched vector
x2 = A2 @ x # rotated vector
x3 = A3 @ x # sheared vector

fig, axs = plt.subplots(figsize=[5.00, 5.00])
axs.grid(linestyle='-', linewidth=2)

# stretched vector
axs.quiver(0, 0, x1[0], x1[1], angles='xy', scale_units='xy', scale=1, color=['green'], label='Stretched')
# original vector
axs.quiver(0,0, x[0] , x[1], angles='xy', scale_units='xy', scale=1, color=['black'], label='Original')
# rotated vector
axs.quiver(0,0, x2[0] , x2[1], angles='xy', scale_units='xy', scale=1, color=['blue'], label='Rotated')
# sheared vector
axs.quiver(0,0, x3[0] , x3[1], angles='xy', scale_units='xy', scale=1, color=['orange'], label='Sheared')

axs.annotate(text=r'$v$', xy=(0, 1), va='top',ha='center', fontsize=20)
axs.annotate(text=r'$v_1$', xy=(2, 3), va='top',ha='center', fontsize=20)
axs.annotate(text=r'$v_2$', xy=(2, 0), va='top',ha='center', fontsize=20)
axs.annotate(text=r'$v_3$', xy=(4.5, 1), va='top',ha='center', fontsize=20)

axs.set_xlabel(r"$x_1$")
axs.set_ylabel(r"$x_2$")
axs.set_xlim([-1,5])
axs.set_ylim([-1,5])
axs.set_title("Linear transformation")
plt.legend(bbox_to_anchor=(1.4, 1))
plt.show()
```

# Eigenvector and eigenvalue

## eigen vector

- def: an eigenvector of a **square matrix** $A \in \mathbb{R}^{n \times n}$ is a nonzero vector $\mathbf{x} \in \mathbb{R}^{n}$ such that, for some scalar $\lambda$:

    $$
    A\mathbf{x} = \lambda x
    $$    

that is, the effect of linear transformation vector $\mathbf{x}$ by matrix $A$ is same as stretching the vector $\mathbf{x}$ by a scalar $\lambda$

- a square matrix may have multiple **eigenvector** and **eigenvalue** (characteristic value) (real or complex),

    each eigenvector $\mathbf{x}$ corresponds to its own eigenvalue $\lambda$

- linear transformations can be characterized by their **invariant space**, or by **unchanged vectors**


- e.g. in a 3D rotation, unchanged vector is axis of rotation


- invariant space: if $A$ is a $3 \times 3$ matrix with a complex eigenvalue, 

    then there is a plane in $\mathbb{R}^3$ on which $A$ acts as rotation axis  (possibly combined with scaling),

    every vector in that plane is rotated into another point on the **same plane**,
    
    the plane is said to be **invariant** under $A$

## eigenspace

- def: eigenspace of a square matrix $A$ corresponding to a specific eigenvalue $\lambda$ is **a set of vectors** that consists of **zero vector and all the eigenvectors** corresponding to the $\lambda$

- i.e. the set of all solutions $\mathbf{x}$ of characteristic equation

$$
(A- \lambda I)\mathbf{x}=\mathbf{0}
$$

- e.g. For a matrix $A=\begin{bmatrix}
1 & 6\\ 
5 & 2
\end{bmatrix}$,

    the eigenspace corresponding to $\lambda=7$ is the line through origin $(0, 0)$ and $(1, 1)$;
    
    the eigenspace corresponding to $\lambda=-4$ is the line through origin $(0, 0)$ and $(6, -5)$;

## eigen decomposition

- def: decomposing a square matrix $A$ into a product of a square matrix $U$ , a diagonal matrix $\Lambda$ and an inverse of the square matrix $U^{-1}$


- equals to change matrix $A$ into a matrix consists of basis of eigenspace -> stretch by multiply diagonal matrix $\Lambda$ -> change back by multiply inverse of matrix

$$
A = U \Lambda U^{-1}=\begin{bmatrix}
| & | & |\\ 
u_1 & u_2 & u_3\\ 
| & | & |
\end{bmatrix}
\begin{bmatrix}
\lambda_1 & 0 & 0\\ 
0 & \lambda_2 & 0\\ 
0 & 0 & \lambda_3
\end{bmatrix}
\begin{bmatrix}
- & u_1 & -\\ 
- & u_2 & -\\ 
- & u_3 & -
\end{bmatrix}
$$

where $U \in \mathbb{R}^{n \times n}$ is a square matrix that columns are $n$ linearly independent eigenvectors of $A$

$\Lambda \in \mathbb{R}^{n \times n}$ is a diagonal matrix, the diagonal entries are eigenvalues of $A$, corresponding to eigenvectors in $U$

- eigen decomposition can be derived from fundamental property of eigen vectors

$$
A \mathbf{x} = \lambda \mathbf{x} \\[1em]
AU = U\Lambda \\[1em]
A = U \Lambda U^{-1}
$$

### diagonalization theorem

- all matrix can be SVDed but not all matrix can be eigen decomposed

- diagonalization theorem

    an $n \times n$ matrix $A$ is diagonalizable if and only if it has $n$ linearly independent eigenvectors
    

- i.e., $A$ is diagonalizable if and only if there is enough eigenvectors to form a basis of  $\mathbb{R}^{n }$ (**eigenvector basis**)


- i.e., $A$ is diagonalizable if for each eigenvalue, the dim of its eigenspace = multiplicity of the eigenvalue. 

    e.g., if a eigenvalue has multiplicity 2, then it should has 2 different eigenvectors


- undiagonalizable matrix: multiplicities in eigenvalue $\neq$ multiplicity in eigenvectors

- e.g. let $A=\begin{bmatrix}
1 & 1\\ 
0 & 1
\end{bmatrix}$, this matrix has eigenvalue 1 with multiplicity 2, but only has 1 eigenvector $\begin{bmatrix}
0\\ 
1
\end{bmatrix}$

$$
p(\lambda)=(\lambda-1)^2=0
$$



### special case: spectral theorem

- spectrum of a square matrix $A$: the set of eigenvalues of $A$

$$
\text{Spectrum} = \left\{\lambda_1, ..., \lambda_n\right\}
$$


- spectral theorem: **a n x n symmetric matrix** $A$ has following properties:

    1. $A$ has n real eigenvalues, counting multiplicities
    
    2. dim of the eigenspace for each eigenvalue $\lambda$ = multiplicity of $\lambda$ as a root of characteristic equation
    
    3. eigenspaces are mutually orthogonal, coz eigenvectors corresponding to different eigenvalues are orthogonal
    
    4. $A$ is orthogonally diagonalizable

- since $U$ is an orthogonal matrix, then $U^{-1}= U^T$

$$
A = U \Lambda U^{-1}= U \Lambda U^T
$$



# Methods to find eigen vectors/values

- method 1: solving characteristic equation


- method 2: power iteration


- method 3: Rayleigh quotient


- method 4: `numpy.linalg.eig`

## characteristic equation

- def: characteristic equation is the scalar equation 

$$
\text{det}(A-\lambda I)=0
$$

where $\text{det}$ is the determinant of matrix $A-\lambda I$


- a scalar $\lambda$ is an eigenvalue of a square matrix $A$ if and only if it satisfies the characteristic equation

    i.e. **$\lambda$ is root of characteristic equation**

- determinant: $\text{det}(A)$ or $|A|$ is a scalar

    determinant of an $n \times n$ matrix $A=[a_{ij}]$ is summation of n terms like $\pm a_{1j}\text{det}A_{1j}$, with plus and minus sign occur alternatively

### characteristic polynomial

- characteristic polynomial: if $A$ is an $n \times n$ matrix, then determinant is a **polynomial of degree $n$**

$$
p(\lambda)=\text{det}(A-\lambda I)
$$

- **multiplicity of an eigenvalue** $\lambda$ is its multiplicity as a root of the characteristic equation

- e.g., for a $4 \times 4$ matrix 

$$
A=\begin{bmatrix}
5 & -2  & 6 & -1\\ 
0 & 3 & -8 & 0\\ 
0 & 0 & 5 & 4\\ 
0 & 0 & 0 & 1
\end{bmatrix}
$$

the characteristic equation is 

$$
(\lambda -5)^{\color{Red} 2 }(\lambda -3)(\lambda -1)=0
$$

the characteristic polynomial of degree $n=4$ is

$$
\lambda^4 - 14 \lambda^3 + 68 \lambda^2  - 130 \lambda +75
$$

the eigenvalue $\lambda = 5$ has **multiplicity 2**

### finding eigenvector

- after getting eigenvalue $\lambda$ from characteristic equation, plug in $\lambda$ back to 
    
    $$
    (\lambda I-A)\mathbf{x}=0
    $$
    
    use row operations to find eigenvector $\mathbf{x}$

- e.g. For a matrix $A=\begin{bmatrix}
1 & 6\\ 
5 & 2
\end{bmatrix}$,

    first eigenvalue $\lambda_1=7$, eigenvector $\mathbf{x}_1=(1, 1)$

    second eigenvalue $\lambda_2=-4$, eigenvector $\mathbf{x}_2=(6, -5)$

$$
A-\lambda I = \begin{bmatrix}
1-\lambda & 6\\ 
5 & 2 - \lambda
\end{bmatrix}
$$

$
p(\lambda)=\text{det}(A-\lambda I)= (1-\lambda)(2 - \lambda)-5 \times 6 = \lambda^2-3\lambda -28=(\lambda -7)(\lambda +4)=0
$

$
\Rightarrow \lambda_1 = 7, \lambda_2 = -4
$

 Plug $\lambda_1$ back to $(\lambda I A)\mathbf{x}=0$

$$
\lambda I A = 7I-A=\begin{bmatrix}
6 & -6\\ 
-5 & 5
\end{bmatrix}
=\begin{bmatrix}
1 & -1\\ 
0 & 0
\end{bmatrix}
$$

$$
\begin{bmatrix}
1 & -1\\ 
0 & 0
\end{bmatrix}
\begin{bmatrix}
x\\ 
y
\end{bmatrix}
=\begin{bmatrix}
0\\ 
0
\end{bmatrix} \Rightarrow x-y=0 
$$

the general solution has the form $a \begin{bmatrix}
1\\ 
1
\end{bmatrix} (a \neq 0)$

 Plug $\lambda_2$ back to $(\lambda I A)\mathbf{x}=0$

$$
\lambda I A = -4I-A=\begin{bmatrix}
-5 & -6\\ 
-5 & -6
\end{bmatrix}
=\begin{bmatrix}
5 & 6\\ 
0 & 0
\end{bmatrix}
$$

$$
\begin{bmatrix}
5 & 6\\ 
0 & 0
\end{bmatrix}
\begin{bmatrix}
x\\ 
y
\end{bmatrix}
=\begin{bmatrix}
0\\ 
0
\end{bmatrix} \Rightarrow 5x+6y=0 
$$

the general solution has the form $b \begin{bmatrix}
6\\ 
-5
\end{bmatrix} (b \neq 0)$

## Power iteration method

- in application, we only need a close numerical approximation of eigenvalues

### algorithm

1. randomly select an initial vector $\mathbf{x}_0$ whose largest entry is 1


2. for $k=0, 1, ...,$

    a. compute $A\mathbf{x}_k$
    
    b. let $\mu_k$ be the entry in $A\mathbf{x}_k$ whose absolute value is largest
    
    c. compute $\mathbf{x}_{k+1} = \frac{1}{\mu_k}A\mathbf{x}_k$


3. for almost all choices of $\mathbf{x}_0$,

    the seq $\left \{ \mu_k \right \}$ converges to **dominant eigenvalue** $\lambda_1$,
   
    the seq $\left \{ \mathbf{x}_k \right \}$ converges to a corresponding eigenvector $\mathbf{v}_1$

### e.g.

- let $A = \begin{bmatrix}
6 & 5\\ 
1 & 2
\end{bmatrix}$,

    start with a random vector $\mathbf{x}_0=\begin{bmatrix}
0\\ 
1
\end{bmatrix}$,

    iterates for $k=5$,
    
    - the seq $\left \{ \mathbf{x}_k \right \}$ converges to eigenvector $\begin{bmatrix}
1\\ 
0.2
\end{bmatrix}$,
    
    - the seq $\left \{ A\mathbf{x}_k \right \}$ converges to 7 times of eigenvector $\begin{bmatrix}
7\\ 
1.4
\end{bmatrix}$,
    
    - the seq $\left \{ \mu_k \right \}$ converges to dominant eigenvector $7$

- the seq converges so quickly only with 5 iterations 

    coz the second eigenvalue of $A$ is much smaller than the dominant eigenvalue $\lambda_2=1 << \lambda_1=7$

### explanation

- applied condition: an $n \times n$ diagonalizable matrix $A$ has a **strictly dominant eigenvalue** $\lambda_1$

$$
{\color{Red}{\left | \lambda_1 \right |} } >> \left | \lambda_2 \right | \geq \left | \lambda_3 \right | \geq ... \geq \left | \lambda_n \right | 
$$

i.e. $\lambda_1$ is larger in absolute value than other eigenvalues

- the rate of convergence depends on ratio $\left | \frac{\lambda_2}{\lambda_1} \right |$

    if $|\lambda_1|$ is not greatly larger than $|\lambda_2|$, then ratio $\left | \frac{\lambda_2}{\lambda_1} \right | \approx 1$,

    the seq $\left \{ \mu_k \right \}$ and $\left \{ \mathbf{x}_k \right \}$ may converge very slowly,

    in this case, other approximation methods maybe preferred.

since eigenvectors $\mathbf{v}_1, ..., \mathbf{v}_n$ is a set of basis of $\mathbb{R}^n$,

for any vector $\mathbf{x}\in \mathbb{R}^n$, it can be written as linear combination of that basis eigenvectors

$$
\mathbf{x}=c_1 \mathbf{v}_1 + ...+ c_n \mathbf{v}_n
$$

during $k$ iterations ($k=1, 2, ...$)

multiply $\mathbf{x}$ by $A^k$, we have

$$
A^k \mathbf{x} = c_1 A^k \mathbf{v}_1 + ...+ c_n  A^k \mathbf{v}_n
$$

$$
A^k \mathbf{x} = c_1 (\lambda_1)^k \mathbf{v}_1 + ...+ c_n  (\lambda_n)^k \mathbf{v}_n
$$

assume $c_1 \neq 0$, dividing by $(\lambda_1)^k$, we have

$$
\mathbf{x}_k=\frac{A^k \mathbf{x}}{(\lambda_1)^k} = c_1 \mathbf{v}_1 + c_2 \left(\frac{\lambda_2}{\lambda_1}\right)^k \mathbf{v}_2...+ c_n \left(\frac{\lambda_n}{\lambda_1}\right)^k \mathbf{v}_n
$$

this scaling is used to make them converge to $c_1 \mathbf{v}_1$

since $
{\color{Red}{\left | \lambda_1 \right |} } >> \left | \lambda_2 \right | \geq \left | \lambda_3 \right | \geq ... \geq \left | \lambda_n \right | 
$,

$\left(\frac{\lambda_2}{\lambda_1}\right)^k \rightarrow 0 ,..., \left(\frac{\lambda_n}{\lambda_1}\right)^k \rightarrow 0$,

hence, for large $k$, a scalar multiple of $A^k \mathbf{x}$: $\frac{A^k \mathbf{x}}{(\lambda_1)^k}$ points in almost **same direction as $c_1 \mathbf{v}_1$**

$$
\frac{A^k \mathbf{x}}{(\lambda_1)^k} \xrightarrow{k \rightarrow 0} c_1 \mathbf{v}_1
$$

then $A^k \mathbf{x}$ also points in almost same direction of eigenvector $\mathbf{v}_1$

- in geometry, as $k \rightarrow \infty$, the angle between eigenvector $\mathbf{v}_1$ and one of element in seq $\left \{\mathbf{x}, A \mathbf{x}, A^2 \mathbf{x},...,A^k \mathbf{x} \right \}\rightarrow 0$,
    
    so the last element $A^k \mathbf{x}$ is a good approximation of true eigenvector $\mathbf{v}_1$

- e.g. let $A = \begin{bmatrix}
1.8 & 0.8 \\ 
0.2 & 1.2
\end{bmatrix}$,

    we can compute $A$ has eigenvalue 2 and 1,
    
    the eigenspace for the dominant eigenvalue $\lambda_1=2$ is the line through $\mathbf{0}$ and $\mathbf{v}_1=\begin{bmatrix}
4\\ 
1
\end{bmatrix}$

    randomly pick a vector $\mathbf{x}=\begin{bmatrix}
-0.5\\ 
1
\end{bmatrix}$,

    compute $A^k\mathbf{x}$ for $k=0,1,...8$,
    
    we can see in figure that $A^8\mathbf{x}=\begin{bmatrix}
101.5\\ 
26.5
\end{bmatrix}$ is approaching eigenvector $\mathbf{v}_1=\begin{bmatrix}
4\\ 
1
\end{bmatrix}$

- Note: the scaling made by dividing ${(\lambda_1)}^k$ is used to make equation converge to $c_1 \mathbf{v}_1$, 

    in practical, we can't do that coz we don't know ${\lambda_1}$,
    
    but we can scale each $A^k\mathbf{x}$ to be $\mathbf{x}_k$ that its largest entry is 1,
    
    then the resulting seq $\left \{ \mathbf{x}_k \right \}$ will converge to a multiple of $\mathbf{v}_1$ whose largest entry is 1

### implementation 1 by Smita

Smita choose a different scaling factor $\left \| A \mathbf{b}_k \right \|$ rather than in textbook the largest entry in $A \mathbf{b}_k$

1. randomly select an initial vector $\mathbf{b}_0$ whose largest entry is 1


2. for $k=0, 1, ...,$

    a. compute $A\mathbf{b}_k$
    
    b. compute $\mathbf{b}_{k+1} = \frac{1}{\left \| A \mathbf{b}_k \right \|}A\mathbf{b}_k$


3. for almost all choices of $\mathbf{b}_0$,
   
    the seq $\left \{ \mathbf{b}_k \right \}$ converges to a corresponding eigenvector $\mathbf{v}_1$

```python
def power_iteration(A: np.ndarray, num_iter: int) -> np.ndarray:
    """
    A: a square matrix who has a dominant eigenvalue
    num_iter: number of iterations

    return: (first eigenvector of A, first eigenvalue of A)
    """
    # randomly choose a vector to decrease the chance that it is orthogonal to eigenvector of A 
    bk = np.random.rand(A.shape[1])
    plt.figure(figsize=(4, 4))
    plt.grid(linestyle='-', linewidth=1)
    plt.xlim([-0.5,1.5])
    plt.ylim([-0.5,0.75]) 

    for i in range(num_iter):

        Ab = A @ bk  

        bk = Ab / np.linalg.norm(Ab) # normalize the vector
        
        # plot each b_k vector
        plt.quiver(0, 0, bk[0] , bk[1], angles='xy', scale_units='xy', scale=1, color=(0.2+(0.1*i), 0.5, 0.5)) 

    return bk, np.linalg.norm(A @ bk)

A=np.array([[3, 3],[0, 2]])
power_iteration(A, 6)
```

- let $A=\begin{bmatrix}
3 & 3\\ 
0 & 2
\end{bmatrix}$, after 6 iterations, vector $\mathbf{b}_k$ converges to eigen vector $\begin{bmatrix}
1 & 0
\end{bmatrix}$, corresponding dominant eigenvalue is $\lambda = 3$

### implementation 2

From [ML Wiki Power Iteration](http://mlwiki.org/index.php/Power_Iteration)

```python
def eigenvalue(A, v):
    Av = A.dot(v)
    return v.dot(Av)

def power_iteration(A, threshold=0.01):
    n, d = A.shape

    v = np.ones(d) / np.sqrt(d) # initialize a vector
    ev = eigenvalue(A, v)

    while True:
        Av = A @ v 
        v_new = Av / np.linalg.norm(Av)

        ev_new = eigenvalue(A, v_new)
        
        # if eigenvalue no longer update, stop iteartion
        if np.abs(ev - ev_new) < threshold:
            break

        v = v_new
        ev = ev_new
        
    return ev_new, v_new
```

```python
A=np.array([[3, 3],[0, 2]])
power_iteration(A)
```

## Rayleigh quotient: find eigenvalues

- Rayleigh quotient:

$$
R(\mathbf{u}) = \frac{\mathbf{u}^T A \mathbf{u}}{\mathbf{u}^T  \mathbf{u}}=\lambda
$$

- derivation

    $$
    A \mathbf{u} = \lambda \mathbf{u}
    $$
    
    $$
    \mathbf{u}^T (A \mathbf{u}) = \mathbf{u}^T(\lambda \mathbf{u}) = \lambda (\mathbf{u}^T \mathbf{u})
    $$

- if $\mathbf{u}$ is an estimated eigenvector for $\lambda$, plug in $\mathbf{u}$ to find the eigenvalue $\lambda$
    

- when $A$ is a symmetric matrix ($A^T=A$),

    Rayleigh quotient will have twice as many digits of accuracy as scaling factor $\mu_k$ in power method

### Next eigenvector

- take out the largest eigenvector $\mathbf{u}$ from the matrix $A$

$$
B = A - \lambda \mathbf{u}\mathbf{u}^T
$$

then the second eigenvalue of $A$ is the largest eigenvalue of matrix $B$

## Numpy/Scipy API

`numpy.linalg.eig` or `scipy.linalg.eig`: 

- used for general square matrix

- uses a slower algorithm

- return unsorted eigenvalues. `numpy.linalg.eigvals`: only return eigenvalues of a general square matrix


`numpy.linalg.eigh`: 

- used for symmatrix matrix 

- uses a faster algorithm that takes advantage of matrix is symmetric without checking, calculate eigenvectors just by lower triangular part of the matrix.

- return sorted eigenvalues. `numpy.linalg.eigvalsh`: only return eigenvalues of a symmatrix matrix 

`numpy.linalg.eig(A)`

param: `A`: a matrix for which eigenvalues and eigenvectors to be computed

return:

- `w`: an array of eigenvalues, 

    each repeated according to its multiplicity, 
    
    but **not ordered according to size**

- `v`: an array of **normalized** unit eigenvectors,

    ith column of `v` (`v[:,i]`) is the eigenvector corresponding to the ith eigenvalue `w[i]`

- e.g., diagonal real matrix $A$ with real eigenvalues and real eigenvectors

```python
A=np.array([[3, 3],[0, 2]])
np.linalg.eig(A)
```

```python
A = np.diag((1, 2, 3))

e_values, e_vectors = np.linalg.eig(A)

e_values, e_vectors
```

# measure of spread

- measure of spread: variance, entropy

## variance

- variance of a random scalar-valued variable $X \in \mathbb{R}$


$$
\sigma^2_X=Var(X) = E[(X-E[X])^2]
$$

- sample variance is **biased**

$$
E[\sigma^2_X]=\frac{n-1}{n}\sigma^2 < \sigma^2
$$

where $n$ is sample size

$$
\sigma^2_X=\frac{1}{n}\sum_{i=1}^n (X_i - \bar X)^2
$$


- corrected sample variance is **unbiased**

$$
s^2 = \frac{n}{n-1} \sigma^2_X=\frac{1}{n-1}\sum_{i=1}^n (X_i - \bar X)^2
$$


- similarly, corrected unbiased sample covariance

$$
q_{jk} =\frac{1}{n-1}\sum_{i=1}^n (X_{ij} - \bar X_j)(X_{ik} - \bar X_k)
$$

## covariance

- covariance: measure joint variability


- covariance of 2 1D random variable $X,Y \in \mathbb{R}$ is a scalar


$$
Cov(X,Y) = E[(X-E[X])(Y-E[Y])]\\[1em]
= 0\ (\text{X, Y are independent})
$$

- **variance-covariance matrix** of a random vector $\mathbf{a} \in \mathbb{R}^p$ is a square matrix $ \in \mathbb{R}^{p \times p}$

    diagonal entries are $Var(\mathbf{a}_i)$

$$
Cov(\mathbf{a},\mathbf{a})= Var(\mathbf{a})
=E[\left(\mathbf{a}-E[\mathbf{a}]\right)\left(\mathbf{a}-E[\mathbf{a}]\right)^T]\\[1em]

=\begin{bmatrix}
Cov(a_1, a_1) & ... & Cov(a_1, a_n)\\ 
\vdots  & \ddots  & \vdots \\ 
Cov(a_n, a_1) & ... & Cov(a_n, a_n)
\end{bmatrix}\\[1em]
=\begin{bmatrix}
Var(a_1) & ... & Cov(a_1, a_n)\\ 
\vdots  & \ddots  & \vdots \\ 
Cov(a_n, a_1) & ... & Var(a_n)
\end{bmatrix}
$$

- Note: for matrix $\mathbf{X} \in \mathbb{R}^{p \times n}$ with $n$ samples and $p$ features is a square matrix $\in \mathbb{R}^{p \times p}$, covariance is $XX^T/n$

- **population covariance matrix** of a matrix $\mathbf{X} \in \mathbb{R}^{n \times p}$ with $n$ samples and $p$ features is a square matrix $\in \mathbb{R}^{p \times p}$

$$
Cov(\mathbf{X})=\frac{\left[\mathbf{X}- \mu\right]^T \left[\mathbf{X}- \mu\right]}{n}
$$

where $\mu=\frac{1}{m}\mathbf{1}^T\mathbf{X} \in \mathbb{R}^{p}$ is column mean of matrix $\mathbf{X}$ known as a **priori**

when all column mean are 0, $\mu=\mathbf{0}$,

$$
Cov(\mathbf{X})=\frac{1}{n}{\mathbf{X}}^T \mathbf{X}
$$


- **sample covariance matrix**: estimate column mean from the data, Note **normalization is by $n-1$**

$$
Cov(\mathbf{X})=\frac{\left[\mathbf{X}-\bar{\mathbf{X}}\right]^T \left[\mathbf{X}-\bar{\mathbf{X}}\right]}{n-1}
$$

when mean is 0, $\bar{\mathbf{X}}=\mathbf{0}$,

$$
Cov(\mathbf{X})=\frac{{\mathbf{X}}^T \mathbf{X}}{n-1}
$$



### properties of covariance matrix

- symmetric

    $$
    Cov(X,Y)=Cov(Y,X)
    $$
    

- positive semidefinite

    since any matrix in the form $XX^T$ is positive semidefinite, then covariance matrix is also
    

- all eigenvalues are real positive

### covariance matrix as linear transformation operator

- given uncorrelated data points in a matrix $X$

    - left multiply $X$ by covariance matrix $\Sigma$ of $X$ can convert $X$ to a highly correlated matrix
    
    $$\Sigma X$$
    
    - left multiply $\Sigma X$ by inverse of covariance matrix $\Sigma$ of $X$ can de-correlates variables
    
    $$\Sigma^{-1}(\Sigma X)$$

### covariance matrix as  quadratic form

- for a random variable $\mathbf{X} \in \mathbb{R}^{n \times n}$, vector $\mathbf{a}, \mathbf{b} \in \mathbb{R}^{n}$

    $$
    Cov(\mathbf{a}^T\mathbf{X}, \mathbf{X}) = \mathbf{a}^T\mathbf{\Sigma} \in \mathbb{R}^{1 \times n}
    $$

- covariance between 2 linear combination $\mathbf{a}^T\mathbf{X}, \mathbf{b}^T\mathbf{X}$ of variable $\mathbf{X}$ is **bilinear form**

    $$
    Cov(\mathbf{a}^T\mathbf{X}, \mathbf{b}^T\mathbf{X}) = \mathbf{a}^T\mathbf{\Sigma}\mathbf{b} \in \mathbb{R}
    $$

- variance of a linear combination is covariance with itself
    
    $$
    Var(\mathbf{a}^T\mathbf{X})=Cov(\mathbf{a}^T\mathbf{X}, \mathbf{a}^T\mathbf{X}) = \mathbf{a}^T\mathbf{\Sigma}\mathbf{a} \in \mathbb{R}
    $$ 

- quadratic form: a function $Q: \mathbb{R}^n \rightarrow \mathbb{R}$ with expression of the form
    
    $$
    Q(\mathbf{x}) = \mathbf{x}^T A \mathbf{x}
    $$
    
    where $A$ is an $n \times n$ symmetric matrix, called **matrix of quadratic form**

- A quadratic form $Q= \mathbf{x}^T A \mathbf{x}$ is

    a. positive definite if $Q(\mathbf{x})>0$ for all $\mathbf{x} \neq \mathbf{0}$. if and only if eigenvalue of $A$ are all positive
    
    b. negative definite if $Q(\mathbf{x})<0$ for all $\mathbf{x} \neq \mathbf{0}$. if and only if eigenvalue of $A$ are all negative
    
    c. indefinite if $Q(\mathbf{x})$ assumes both positive and negative values. if and only if eigenvalue of $A$ has both positive and negative values
    
    d. **positive semidefinite** if $Q(\mathbf{x}) \geq 0$ for all $\mathbf{x} \neq \mathbf{0}$. if and only if eigenvalue of $A$ are all non-negative
    
    e. negative semidefinite if $Q(\mathbf{x}) \leq 0$ for all $\mathbf{x} \neq \mathbf{0}$

- Bilinear form : a function $B: \mathbb{R}^n \times  \mathbb{R}^n \rightarrow \mathbb{R}$ that
 
  - $B(\mathbf{a}+\mathbf{b},\mathbf{c})=B(\mathbf{a},\mathbf{c})+B(\mathbf{b},\mathbf{c})$ and $B(\lambda \mathbf{a},\mathbf{b})=\lambda B(\mathbf{a},\mathbf{b})$
 
  - $B(\mathbf{a},\mathbf{b}+\mathbf{c})=B(\mathbf{a},\mathbf{b})+B(\mathbf{a},\mathbf{c})$ and $B(\mathbf{a},\lambda \mathbf{b})=\lambda B(\mathbf{a},\mathbf{b})$

- dot product is a bilinear form

## correlation

- correlation: covariance normalized by standard deviation

$$
corr(X,Y)=\frac{Cov(X,Y)}{\sigma_X \sigma_Y}
$$

# ex1: compute covariance

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def Cov(x: np.ndarray, y: np.ndarray) -> float:
    """Returns covariance of vectors x and y of shape (n, )"""

    return np.sum((x - x.mean())*(y - y.mean()))/(x.shape[0] - 1)

def cov_mat(X: np.ndarray) -> np.ndarray:
    """Returns a covariance matrix (2, 2) of a matrix X of shape (n, 2)"""
    
    return np.array([[Cov(X[0], X[0]), Cov(X[0], X[1])], \
                     [Cov(X[1], X[0]), Cov(X[1], X[1])]])
```

```python
# Normal distributed x and y vector with mean 0 and standard deviation 1
n = 500
x = np.random.normal(0, 10, n) # x ~ N(0, 10)
y = np.random.normal(0, 1, n)+x  # y ~ N(0, 1) and correlate with x
X = np.vstack((x, y)).T # shape: (n, 2)

# Calculate covariance matrix 
np.isclose(cov_mat(X.T), np.cov(X.T))

plt.scatter(X[:, 0], X[:, 1], s=5, color ='blue')
plt.xlabel(r"$X_1$")
plt.ylabel(r"$X_2$")
plt.title('Generated Data')
```

# Cholesky decomposition of covariance matrix

Cholesky decomposition of the covariance matrix to generate specific variables. 

First consider an eigendecomposition of the covariance matrix

$$
\Sigma=USU^{-1}
$$

Here $U$ can be thought of as a rotation matrix, and $S$ as a scaling matrix

If we apply just the rotation $U$ and a modified scaling matrix $\sqrt{S}$ to normal random data $X$, then we will create exactly the covariance values from our covariance matrix. 

This is explained here: http://www.cs.utah.edu/~tch/CS4640F2019/resources/A%20geometric%20interpretation%20of%20the%20covariance%20matrix.pdf 


# geometry of covariance matrix

- we can rewrite covariance matrix $\Sigma \in \mathbb{R}^{p \times p}$ in the form of SVD:

$$
\Sigma = U \Lambda U^{-1} = U LL U^{-1} 
$$

$$
\Sigma = U \Lambda U^{-1}=\begin{bmatrix}
| & | & |\\ 
u_1 & u_2 & u_3\\ 
| & | & |
\end{bmatrix}
\begin{bmatrix}
\lambda_1 & 0 & 0\\ 
0 & \lambda_2 & 0\\ 
0 & 0 & \lambda_3
\end{bmatrix}
\begin{bmatrix}
- & u_1 & -\\ 
- & u_2 & -\\ 
- & u_3 & -
\end{bmatrix}
$$

where $L=\Lambda ^{1/2}$ is a diagonal matrix with diagonal entries the square root ot diagonal matrix $\Lambda$

$$
L=\begin{bmatrix}
\sqrt{\lambda_1} &  & &\\ 
 & \sqrt{\lambda_2} & &\\ 
 &  & \ddots &\\
 &  & & \sqrt{\lambda_n}
\end{bmatrix}
$$

- if apply the matrix $C=UL=U\Lambda ^{1/2} \in \mathbb{R}^{p \times p}$ to a vector $X \sim N(\mathbf{0}, \mathbf{I}) \in \mathbb{R}^{p}$

    the resulted data is rotated scaled data $\tilde X =CX  \sim N(\mathbf{0}, \sigma^2\mathbf{I}) \in \mathbb{R}^{p}$
    

- if apply the matrix $C=UL=U\Lambda ^{1/2}$ to a matrix $X \in \mathbb{R}^{p \times n} \sim N(\mathbf{0}, \mathbf{I_p})$ with n **uncorrelated** samples and p features,

    the resulted data is rotated scaled data $\tilde X =CX   \in \mathbb{R}^{p \times n}  \sim N(\mathbf{0}, \Sigma)$ with covariance matrix $\Sigma$
    

- Note: orthonormal matrix $U$ is for rotation, diagonal matrix $L$ is for stretching

proof $Cov(\tilde X)=\Sigma$


- Covariance matrix of an uncorrelated dataset $X \sim N(\mathbf{0}, \mathbf{I_p}) \in \mathbb{R}^{p \times n}$ consists of p independent normal random variables $\sim N(\mathbf{0},\mathbf{I}) \in \mathbb{R}^{n}$ is an identity matrix $\in \mathbb{R}^{p \times p}$:

$$
I_p = XX^T
$$


- for simplicity, consider another correlated dataset $X_0 \in \mathbb{R}^{p \times m}$ as a mean-centered matrix, so its covariance matrix $\Sigma \in \mathbb{R}^{p \times p}$:

$$
\Sigma = \tilde X \tilde X^T = U \Lambda U^{-1} = U LL U^{-1}
$$


- after apply matrix $C=UL$ to uncorrelated dataset $X$, covariance matrix $\tilde \Sigma$ of new dataset $\tilde X$ becomes same as $\Sigma$:

$$
\tilde \Sigma = \tilde X \tilde X^T = (ULX)(ULX)^T = (ULX)(X^T L^T X^T) \\[1em]
= UL (XX^T) L^T U^T=UL I_n L^T U^T \\[1em]
=ULL^T U^T=ULL U^T =U \Lambda U^T = \Sigma
$$

```python
# generate highly correlated data X0
m = 100
means = np.array([0,0]) # for simplicity, mean-centered
covs = np.array([[1, 0.9],[0.9,1]])  # high correlated
X0= np.random.multivariate_normal(means, covs, m) # (m, 2)
sigma = np.cov(X0.T)
w, U = np.linalg.eig(sigma) # eigendecomposition
L = np.diag(np.sqrt(w)) # square root of diagonal matrix
C = U @ L # transform matrix (2,2)

# generate uncorrelated data X
n = 500
x = np.random.normal(0, 1, n)
y = np.random.normal(0, 1, n)
X = np.vstack((x, y)).T # (n, 2) 
I = np.cov(X.T)

# Correlate: apply C to X 
X_tilde = C @ X.T # (2, n)
sigma_tilde = np.cov(X_tilde)

# Decorrelate: apply C^{-1} to X_tilde
X_prime = np.linalg.inv(C) @ X_tilde
sigma_prime = np.cov(X_prime)

sigma, I, sigma_tilde, sigma_prime, np.isclose(sigma, sigma_tilde), np.isclose(I, sigma_prime)

plt.scatter(X0.T[0], X0.T[1], s=10, color='blue', label=r'Correlated data $X_0$')
plt.scatter(X.T[0], X.T[1], s=10, color='red', label=r'Unorrelated data $X$')
plt.scatter(X_tilde[0], X_tilde[1], color='yellow', s=5, label=r'Transformed data $\tilde{X}$')
plt.scatter(X_prime[0], X_prime[1], s=5, color='green', label=r"DeCorrelated data $X'$")
plt.legend(bbox_to_anchor=(1.1, 1))
plt.xlabel(r"$X_1$")
plt.ylabel(r"$X_2$")
plt.title("Correlate and de-correlate")
plt.show()
```

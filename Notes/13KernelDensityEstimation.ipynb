# kernel density estimation

- for real data as a random variable, we want to estimate probability distribution of data
 
    to build generative/predictive model, estimate mutual information between 2 variables, to capture inherent stochasticity, to reason about outliers and noise.

- 2 ways to estimate pdf of data:

    discrete: draw histogram. bin size matters. problem: have gaps, non-smooth and discontinuous.

    continous: kernel density estimation. smooth and interpolates gaps. bandwidth matters, higher bandwidth: oversmooth, lower bandwidth: undersmooth

- KDE is calculated by integrate over all kernels

$$
\hat f_{h}(x) = \frac{1}{n}\sum_{i=1}^N K_h(x - x_i)
$$

$h$ is bandwidth of kernel function $K(x)$

- mean integrated square error

$$
\text{MISE}(h) = \mathbb{E}[\int (\hat f(x) - f(x))^2 dx]
$$

use Gaussian distribution for actual data distribution $f(x)$

solve for optimal bandwidth 

$$
h = \left(\frac{4 \hat \sigma^5}{3n}\right)^{1/5}
$$
